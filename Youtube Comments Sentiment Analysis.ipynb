# In this project I will perform sentiment analysis on youtube_comments

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading data from csv 

comments=pd.read_csv('C:/Users/27823/Downloads/GBcomments.csv',error_bad_lines=False)
```

```python

#displaying dataframe 
comments.head()
```

```python
from textblob import TextBlob 
```

```python
# checking polarity 

TextBlob("It's more accurate to call it the M+ (1000) be..").sentiment.polarity
```

```python
#Check na values 
comments.isna().sum()
```

```python
#Removing na values
comments.dropna(inplace=True)
```

```python
comments.isna().sum()
```

```python
#finding polarity of all values in comment_text

polarity=[]

for i in comments['comment_text']:
    polarity.append(TextBlob(i).sentiment.polarity)
```

```python
#adding column polarity to comment table 

comments['polarity']=polarity 
```

```python
comments.head(50)
```

```python
#finding postive polarity

comments_postive=comments[comments['polarity']==1]
```

```python
#comments with a polarity of 1
comments_postive
```

```python
#Implementing wordcloud to display postive comments 

from wordcloud import WordCloud,STOPWORDS
```

```python
# Extracting stopwords

stopwords=set(STOPWORDS)
```

```python
stopwords
```

```python
total_comments=' '.join(comments_postive['comment_text'])
```

```python
MyWordCLoud=WordCloud(width=1000,height=500,stopwords=stopwords).generate(total_comments)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(MyWordCLoud)
plt.axis('off')
```

```python
#changing polarity 

comments_negative=comments[comments['polarity']==-1]

total_comments=' '.join(comments_negative['comment_text'])
```

```python
MyWordCLoud=WordCloud(width=1000,height=500,stopwords=stopwords).generate(total_comments)
plt.figure(figsize=(15,5))
plt.imshow(MyWordCLoud)
plt.axis('off')
```
